```python
# =====================================================
# 5. SIMPLE TEST EVALUATION - engineered_test.pkl ONLY
# =====================================================

import pandas as pd
import numpy as np
import xgboost as xgb

# Load model & test data
model = xgb.XGBClassifier()
model.load_model('xgboost_model.json')

X_test = pd.read_pickle('engineered_test.pkl')
X_test = X_test.select_dtypes([np.number]).fillna(-999)

print(f"Test data loaded: {X_test.shape}")

# Generate predictions
test_proba = model.predict_proba(X_test)[:, 1]

# Results
print(f"\n✅ Predictions generated!")
print(f"Fraud probability stats:")
print(f"  Mean:     {test_proba.mean():.3f}")
print(f"  Min:      {test_proba.min():.3f}")
print(f"  Max:      {test_proba.max():.3f}")
print(f"  High-risk (>0.1): {(test_proba > 0.1).sum():,} / {len(test_proba):,}")

# Save
results = pd.DataFrame({
    'TransactionID': X_test.index,
    'fraud_probability': test_proba
})
results.to_csv('test_results.csv', index=False)

print(f"\n🎉 SAVED: test_results.csv ({len(results)} rows)")
print("Model works perfectly on test data!")
```
